```python
import math

x1 = [1.2, 2, 3.8, 4.5]
x2 = [0.5, 4.5, 9.6, 3.4]

def distanciaeuclidiana(x1, x2):
    dim, soma = len(x1), 0
    for i in range(dim):
        soma += math.pow(x1[i] - x2[i], 2)
    return math.sqrt(soma)

print('%.2f' % distanciaeuclidiana(x1, x2))
```

```python
x = boston.data[:, 6:8]
y = boston.target
numrows = len(x)
numcol = len(x[0])
print(numrows)
print(numcol)
print(range(150))
for linha in range(numrows):
    print(x[linha,:] )
```

```python
xt = x[222:, :]
xv = x[0:222, :]
```

```python
yt = y[222:]
yv = y[0:222]
print(y)
```

```python
rowsxt = len(xt)
print(rowsxt)
```

```python
rowsxv = len(xv)
print(rowsxv)
```

```python
import numpy as np
from scipy import stats
k = 5
dist = np.zeros([rowsxt, 1])
yordenado = np.zeros([k, 1])
for linhaxv in range(rowsxv):
    for linhaxt in range(rowsxt):
        dist[linhaxt] = distanciaeuclidiana(xv[linhaxv,:], xt[linhaxt,:])
    #print(dist)
    indices = np.argsort(dist, axis=0)
    #print(indices)
    for i in range(0, k):
        yordenado[i]=yt[indices[i]]
    #print(yordenado)
    m = stats.mode(yordenado) 
    classe = m[0]
    print ("Classe predita %d" % (m[0]))
    print ("Classe real %d" % (yv[linhaxv]))
    #print(ycalculado[0])
```
